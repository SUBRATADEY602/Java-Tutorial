```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
hdata=pd.read_csv('/heart (3).csv')
hdata.head()
```

```python
hdata.shape
```

```python
hdata.describe()
```

```python
hdata['target'].value_counts()
```

```python
x=hdata.drop(columns='target',axis=1)
y=hdata['target']
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_pridiction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_pridiction,y_train)
```

```python
print('accuracy:',training_data_accuracy)
```

```python
x_test_pridiction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_pridiction,y_test)
print('accuracy on test:',test_data_accuracy)
```

```python
input_data=(52,1,0,125,212,0,1,168,0,1,2,2,3)
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)

print(prediction)
if prediction[0] == 0:
    print("You are safe, no need to worry")
else:
    print("Caution, you may suffer from heart disease")
```
